```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import math
```

```python
x=[1,2,3,4,5]
array=np.array(x)
print(array)
y=3*array+5
plt.subplot(2,1,1)
plt.plot(array,y)
amplitude=np.linspace(0,2*math.pi,1000)
plt.subplot(2,1,2)
plt.plot(amplitude,np.sin(amplitude),'-r')
plt.grid()
plt.title("Sine curve")
plt.xlabel("Amplitude")
plt.ylabel("Height")
```
